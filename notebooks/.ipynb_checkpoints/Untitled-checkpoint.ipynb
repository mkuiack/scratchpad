```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import matplotlib
from scipy.optimize import curve_fit


%matplotlib inline 
```

```python
data = pd.read_csv("/home/kuiack/fullcat_fits.csv", names=["slope","slope_err","int","int_err","N_match"])
```

```python
def powlaw(x,a,b):
    return a*x**(b)

x_range = np.linspace(1,50,100)

fit, cov = curve_fit(powlaw, data.N_match, np.sqrt(data.int_err))

plt.plot(data.N_match, np.sqrt(data.int_err),'.')
plt.plot(x_range, powlaw(x_range,fit[0],fit[1]))
# plt.plot(x_range, powlaw(x_range,fit[0],-0.5))
plt.yscale("log")

print fit
```
