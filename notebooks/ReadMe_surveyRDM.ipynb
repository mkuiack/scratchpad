## Raw Data

The observation data is stored in the AARTFAAC archive as calibrated visibilies. The total data volume is nearly 20 TB, far too much to be reproduced here. However, the calibrated visibility data from any observation used in the survey can be made available upon request.


## End-to-End analysis scripts  

AARTFAAC calibrated visibilities can be imaged with 
https://github.com/transientskp/aartfaac-pysquasher

The images are processed for source measurement and transient detection with TraP The LOFAR transient pipeline, v3.1.1
https://github.com/transientskp/tkp/tree/r3.1.1

All methods developed for transient detection can be found in  
https://github.com/mkuiack/scratchpad 


## Figures and Tables


The data files used to generate each figure are stored in surveyRDM. These data can be genereated using survey_sensitivityplot.ipynb. Additional development and experimentation is also presented in the notebook. 

The Trap databases for the candidate sources described in the table, are given in Candidate_databases. 
These are databases which can be accessed with PostgreSQL 9.5.
